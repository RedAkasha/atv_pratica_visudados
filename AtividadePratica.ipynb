**Pergunta**: A anÃ¡lise da covariÃ¢ncia e correlaÃ§Ã£o pode dizer algo sobre a lacuna da quantidade de docentes entre as duas Ã¡reas?

**AÃ§Ã£o**: Verificar estatisticamente a comparaÃ§Ã£o destas duas Ã¡reas no cenÃ¡rio nacional escolar.

```python
plt.style.use('seaborn-v0_8')

cols = ['QT_DOC_BAS', 'CienciasHumanas', 'CienciasNatureza']
means = df[cols].mean()
labels = ['Total docentes', 'Humanas', 'CiÃªncias da Natureza']

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(labels, means, color=['#4c72b0', '#dd8452', '#55a868'])

ax.set_title('ComparaÃ§Ã£o mÃ©dia de docentes: Total de docentes x CiÃªncias Humanas x CiÃªncias da Natureza', fontsize=14, fontweight='bold')
ax.set_ylabel('MÃ©dia de docentes', fontweight='bold')
ax.set_xlabel('Categoria', fontweight='bold')

for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),
                textcoords='offset points',
                ha='center')

plt.tight_layout()
plt.show()
```

**GrÃ¡fico criado**: Barras que representam as mÃ©dias de docentes de cada categoria em destaque.

**PrincÃ­pios de Tufte aplicados**: Apliquei negrito no "label" dos eixos x e y e aumentei a fonte e deixei em negrito o tÃ­tulo principal do grÃ¡fico, para melhor destaque. Usei cores bem distintas nas barras para melhor representaÃ§Ã£o visual.

**Insight visual**: O grÃ¡fico demonstra um destoamento acentuado da mÃ©dia de docentes em ciÃªncias da natureza comparando-se com os de humanas e com o total.

**Descoberta principal**: O cenÃ¡rio nacional de docentes da educaÃ§Ã£o bÃ¡sica carece de profissionais da Ã¡rea de ciÃªncias da natureza (biologia, quÃ­mica e fÃ­sica) com considerÃ¡vel impacto na formaÃ§Ã£o escolar de crianÃ§as e jovens.

**EvidÃªncia**: A mÃ©dia de quantidade de docentes de ciÃªncias da natureza nas escolas Ã© de 1,2; enquanto que a mÃ©dia em ciÃªncias humanas Ã© de 9,9 evidenciando uma lacuna destoante, se compararmos as cargas mÃ©dias de horas estudadas de cada Ã¡rea nos ensinos fundamental e mÃ©dio somados (cerca de 980 horas em ciÃªncias da natureza e 1.140 horas em ciÃªncias humanas).
